```python
#dependent variable ,independent variable
x=dataset.iloc[:,:-1]
print(x)
y=dataset.iloc[:,-1]
print(y)
```

```python
#data normilazation using MinMax
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(x)
print(x_scaled)
```

```python
#splitting the data into traning and testing
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size=0.2,random_state=1)
```

```python
#import the model
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
scores = []
k_Values = []
for i in range(1,50,2):
    model = KNeighborsClassifier(n_neighbors=i)
    model.fit(x_train,y_train)
    preds = model.predict(x_test)
    scores.append(accuracy_score(preds,y_test))
    k_Values.append(i)
print(scores)
print(k_Values)
```

```python
scores = np.array(scores)
max_indx = np.argmax(scores)
max_indx
```

```python
print(scores[7])
print(k_Values[7])
```

```python
plt.plot(k_Values,scores)
```

### accuracy is highest when k = 15

```python
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
final_model = KNeighborsClassifier(n_neighbors=15)
final_model.fit(x_train,y_train)
preds = final_model.predict(x_test)
print(classification_report(preds,y_test))
print(accuracy_score(preds,y_test))
print(confusion_matrix(preds,y_test))
```

```python
sns.heatmap(confusion_matrix(preds,y_test))
```

```python
from sklearn.decomposition import PCA
#Perform PCA to reduce to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(x_scaled)

#Plot the 2D PCA result
plt.scatter(X_pca[:, 0], X_pca[:, 1], c='blue', edgecolor='k')
plt.title('PCA: Reduction from 12D to 2D')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.show()
```
