```python
#model building 
#logistic model
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split as ts
#split the data in train and text
X_train,X_test,y_train,y_test=ts(X_standardized,y,test_size=0.3,random_state=30)
# Create and train the logistic regression model
model = LogisticRegression(solver='lbfgs', verbose=1)
model.fit(X_train, y_train)
```

```python
#accuracy of Logistic Regression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix


y_pred = model.predict(X_test)

# performance
accuracy = accuracy_score(y_test, y_pred)
aclg=accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
```

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['Sex'], df['Survived'])
chi2, p, dof, expected = chi2_contingency(contingency_table)
print("Chi-square test statistic:", chi2)
print("P-value:", p) #accept the ho !!
```

```python

alpha = 0.05  # Set the significance level
if p <= alpha:
    print("reject the null hypothesis associated.")
else:
    print("accept null hypothesis the variables sex and survived are independent.")
```
