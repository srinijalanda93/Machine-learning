```python
# outlier detection technique.
#to check for the outlier consider only the numerical data for that consider int64,float64
number_cols=df.select_dtypes(include=['float64','int64']).columns
# print(number_cols)
# df[number_cols] 
#Data points that fall outside the range defined by mean ± 3 * stdDev are flagged as outliers.
for cols in number_cols:
     mean=df[cols].mean()
     stdDev=df[cols].std()
     outlier=df[(df[cols]<mean-3*stdDev) | (df[cols]>mean+3*stdDev)]
     print(f"Outliers of {cols} : {len(outlier)}")
    
```

```python
#time to do dimensionality reduction
from sklearn.decomposition import PCA
pca=PCA()
pca.fit(df_scled)
#eigne values
ev=pca.explained_variance_
print(ev)
```

```python
#apply pca
pca=PCA(n_components=9)
p_c=pca.fit_transform(df_scled)
print("the pc for some components:",p_c)
#convert into dataframe
pca_df=pd.DataFrame(data=p_c,columns=['i','am','done','with','ml','hope','it','makes','me'])
print(pca_df)
```
