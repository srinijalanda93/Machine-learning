```python
#binning technique
#uisng the pd.cut(df[x],bins='',labels='')
#let create bins and labels (tags)
Bins = [0, 18, 26, 36, 46, float('inf')]
Tags=['under18','18-25','26-35','36-45','46+']
#using cut()
df['GroupedAge']=pd.cut(df['Age'],bins=Bins,labels=Tags)
print(df)
```

```python
'''This will make it easier to visualize and analyze the data 
we can then use a bar plot to show the distribution of customers across these age groups'''

#using matplotlib,seaborn
import matplotlib.pyplot as mplt
import seaborn as sns
```

```python
#count of groupedAge
groupedAge=df['GroupedAge'].value_counts();
groupedAge
groupedAge.index
```

```python
#bar plot
# Create the bar plot
mplt.figure(figsize=(10, 6))
groupedAge.plot(kind='bar', color=['pink','blue'])
mplt.title("The distribution of grouped Age of Customer")
# label the x and y axis
mplt.xlabel('Grouped Age')
mplt.ylabel('no of customer')
mplt.show()
```

#using the cov method
#The covariance matrix will help us understand how the two attributes change together
Values close to 1 indicate a strong positive correlation.
Values close to -1 indicate a strong negative correlation.
Values close to 0 indicate no correlation.

```python
#The range for covariance is from negative infinity to positive infinity.
matrix=df[['Age','Review Rating']]
cov_matrix=matrix.cov()
print(cov_matrix)
```

after we've calculated the covariance matrix we can visualize it using a heat map. 
The heat map will help us identify patterns or trends between these attributes.

```python
# create the heatmap
mplt.figure(figsize=(8, 6))
sns.heatmap(cov_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
mplt.title('covariance matrix')
mplt.show()
```

```python
#Correlation values range from -1 to 1
corr_matrix=matrix.corr();
print(corr_matrix)
```

```python
# create the heatmap
mplt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
mplt.title('covariance matrix')
mplt.show()
```
